### Which customer profiles (job, education, etc.) have higher subscription rates?

```python
a=df.groupby('job')['y'].value_counts(normalize=True).unstack()
```

```python
a.plot(kind='bar',figsize=(10,5),color=['black','maroon'])
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.title('Subscription rate by job', fontsize=14, fontweight='bold')
plt.xlabel('job', fontsize=12)
plt.ylabel('Subsciption', fontsize=12)
plt.show()
```

```python
b=df.groupby('education')['y'].value_counts(normalize=True).unstack()
```

```python
b.plot(kind='bar',figsize=(10,5),color=['black','maroon'])
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.title('Subscription rate by Education', fontsize=14, fontweight='bold')
plt.xlabel('Education', fontsize=12)
plt.ylabel('Subsciption', fontsize=12)
plt.show()
```

### How do job, marital status, and education affect subscription decisions?

```python
g=df.groupby('marital')['y'].value_counts(normalize=True).unstack()*100
```

```python
g.plot(kind='bar',figsize=(10,5),color=['black','maroon'])
plt.legend(['No (0)','Yes (1)'],title='Subscription',bbox_to_anchor=(1.0,1.02))
plt.title('Subscription rate by marital', fontsize=14, fontweight='bold')
plt.xlabel('marital', fontsize=12)
plt.ylabel('Subsciption', fontsize=12)
plt.show()
```

### Does the duration of the call increase the chances of a successful subscription?

```python
plt.figure(figsize=(8, 5))
sb.boxplot(x='y', y='duration', data=df, hue='y', palette=['blue','orange'])
plt.title('Box Plot of Duration by Subscription Outcome (y)')
plt.xlabel('Subscription Outcome (y)')
plt.ylabel('Duration')
plt.show()
```

```python
# Scatter Plot (with jitter)
plt.figure(figsize=(8, 5))
sb.stripplot(x='y', y='duration', data=df, jitter=True, alpha=0.5, hue='y', palette=['blue','orange'])
plt.title('Scatter Plot of Duration vs Subscription Outcome (y)')
plt.xlabel('Subscription Outcome (y)')
plt.ylabel('Duration')
plt.show()
```

```python
average_duration=df.groupby('y')['duration'].mean()
```

```python
average_duration
```

```python
average_duration.plot(kind='bar',figsize=(10,5),color= ['blue','orange'])
plt.legend(['No (0)','Yes (1)'],title='Average duration',bbox_to_anchor=(1.0,1.02))
plt.title('Average duration', fontsize=14, fontweight='bold')
plt.xlabel('Duration', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.show()
```
